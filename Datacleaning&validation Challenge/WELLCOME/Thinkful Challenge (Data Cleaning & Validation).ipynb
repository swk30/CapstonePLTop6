## Setting Up the Code

```python
import pandas as pd
wellcome=pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv', encoding='iso-8859-1')
wellcome.columns=['id', 'publisher','journal', 'article', 'cost']
wellcome.head()
#dataset loaded
#columns are renamed
#top five rows shown below
```

```python
type(wellcome['cost'][1])
```

```python
id_num=wellcome['id']
publisher=wellcome['publisher']
journal=wellcome['journal']
article=wellcome['article']
cost=wellcome['cost']
#new variables assigned for columns
wellcome['cost']=cost.str.strip('£')
#take out pound sign 
wellcome.describe()
```

```python
wellcome['journal']=journal.str.upper()
#the journal names are now in all caps
wellcome['journal']=journal.str.strip()
#removes whitespace
wellcome['journal']=journal.str.strip('.')
#removes periods
wellcome.describe()
```
