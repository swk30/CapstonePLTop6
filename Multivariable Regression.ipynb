```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
excel_file='table_8_offenses_known_to_law_enforcement_new_york_by_city_2013.xls'
df=pd.read_excel(excel_file)
df.head()
```

```python
df.shape
```

```python
#We're gonna begin by cleaning up the top 4 rows + header.
columns=['City','Population','Violent Crime','Murder/Manslaughter','Rape','Rape Legacy','Robbery','Agg Assault',
           'Property Crime','Burgulary','Larceny','Motor_Vehicle Theft','Arson']

df.columns=columns
df[columns]=df[columns].replace(',', '', regex=True)
df.drop(df.head(4).index, inplace=True)
df_new=df.drop(['Rape', 'Arson'], axis=1)
df_new.head()
```

```python
df_new.dropna()
```

```python
#Setting up the specification: Propertycrime=Î±+Population+Population2+Murder+Robbery
df_new['Population^2']=df_new['Population'].apply(float)**2
df_new['Murder_new']=np.where(df_new['Murder/Manslaughter'] > 0, 1, 0)
df_new['Robbery_new']=np.where(df_new['Robbery'].apply(float) > 0, 1, 0)
df_new.head()
```

```python
#New df with just columns Population, Pop^2, Murder and Robbery (as revised)
df_new2=df_new[['City','Population','Population^2','Murder_new','Robbery_new']].copy()
df_new2.head()
```
